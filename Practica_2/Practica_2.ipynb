```python
def impseq(n0,n1,n2):
    # Genera x(n) = delta(n-n0); n1 <= n <= n2
    # ----------------------------------------------
    n=np.arange(n1,n2+1) #Vector de muestras
    x=(n-n0)==0
    return [x,n]
```

```python
def rampa(n0,n1,n2,m):
    n=np.arange(n1,n2+1) #Vector de muestras
    u=(n-n0)>=0
    w=(n-n0)*u
    wp=w*m #Pendiente
    return wp
```

```python
#Punto 3(c)
def seno(freq,theta):
    n=np.arange(0,101)
    w=2*np.pi*freq
    x=np.sin(w+theta*n)    
    return(x)
def expo(sigma,jw0):
    n=np.arange(0,101)
    x=np.exp((sigma+jw0)*n)
    return(x)
eq1=seno(1,0)
eq2=expo(0.03,0)
x3=eq1*eq2
print(x3)
```

```python
#Punto 3(d)
def ramp2(n0,n1,n2):
    n=np.arange(n1,n2+1) #Vector de muestras
    x=np.ones(len(n),int)
    for i in n:
        if n[i]<n0:
            x[i]=0
        if n[i]>=n0:
            x[i]=n[i]-n0
    return x
def escuni2(n0,n1,n2):
    n=np.arange(n1,n2+1) #Vector de muestras
    x=(n-n0)>=0
    x=x*1
    return x
n1=-10
n2=10
eq1=2*ramp2(-3,n1,n2)
eq2=ramp2(2,n1,n2)
eq3=5*escuni2(3,n1,n2)
x4=eq1-eq2-eq3
print(x4)
```

```python
#Punto 4
def rampa(n0,n1,n2,m):
    n=np.arange(n1,n2+1) #Vector de muestras
    u=(n-n0)>=0
    w=(n-n0)*u
    wp=w*m #Pendiente
    return wp,n
[t1,l1]=rampa(0,0,5,1)
[t2,l2]=rampa(5,6,11,1)
[t3,l3]=rampa(10,12,17,1)
rango=np.arange(0,17)
der1=np.diff(t1)
der2=np.diff(t2)
der3=np.diff(t3)
plt.figure(figsize=(10,8))
plt.subplot(2,1,1)
plt.stem(l1,t1,linefmt='black')
plt.stem(l2,t2,linefmt='blue')
plt.stem(l3,t3,linefmt='red')
plt.title('Rampas Originales')
plt.grid()
plt.xlabel('n')
plt.ylabel('r[n]') 
plt.subplot(2,1,2)
plt.stem(l1[1:],der1,linefmt='black')
plt.stem(l2[1:],der2,linefmt='blue')
plt.stem(l3[1:],der3,linefmt='red')
plt.title('Derivadas de las Rampas')
plt.xlabel('n')
plt.ylabel('Derivada')  
plt.grid()
plt.show()
```
