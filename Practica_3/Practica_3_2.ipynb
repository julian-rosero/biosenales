```python
def rutas_pacientes(ruta):
    r_c=[]
    r_p=[]
    r_control=ruta+'\\control\\'
    r_control=os.listdir(r_control)
    r_park=ruta+'\\parkinson\\'
    r_park=os.listdir(r_park)
    for i in range(len(r_control)):
        r_c.append(ruta+'\\'+'control'+'\\'+r_control[i])
    for k in range(len(r_park)):
        r_p.append(ruta+'\\'+'parkinson'+'\\'+r_park[k])
    return r_c,r_p   
```

```python
def senal_continua(matriz,sensores,puntos,epocas):
    s_c=np.reshape(matriz,(sensores,puntos*epocas),order='F')
    return s_c
def energia_prom(señal):
    energia_epoca=np.sum(señal**2)
    energia_prom_canal=np.mean(energia_epoca)
    return energia_prom_canal
def energia_prom_continua(senalcontinua):
    energia=[]
    for i in range(len(senalcontinua)):
        energia_canal=senalcontinua[i]
        valor=energia_prom(energia_canal)
        energia.append(valor)
    return energia
```

```python
def analizar_paciente(ruta_c,ruta_p):
    df_c={}
    df_p={}
    for i in range(len(ruta_c)):
        datos_control=sio.loadmat(ruta_c[i])
        control=datos_control['data']
        senal_control=senal_continua(control,control.shape[0],control.shape[1],control.shape[2])
        eng_c=energia_prom_continua(senal_control)
        df_c['Paciente {}'.format(i+1)]=eng_c
    for k in range(len(ruta_p)):
        datos_parkinson=sio.loadmat(ruta_p[k])
        parkinson=datos_parkinson['data']
        senal_parkinson=senal_continua(parkinson,parkinson.shape[0],parkinson.shape[1],parkinson.shape[2])
        eng_p=energia_prom_continua(senal_parkinson)
        df_p['Paciente {}'.format(k+1)]=eng_p
    return df_c,df_p
c,p=rutas_pacientes('.')
data_c,data_p=analizar_paciente(c,p)
data_c=pd.DataFrame(data_c)
data_p=pd.DataFrame(data_p)
print("Energía Promedio por Canales (Pacientes Control)")
display(data_c)
print("Energía Promedio por Canales (Pacientes Parkinson)")
display(data_p)
```

# Supuestos de pruebas Estadísticas
## Prueba de Levene:
### H[0] - Los datos tienen homocedastisidad (varianzas similares).
### H[1] - Los datos presentan heterocedastisidad (diferencia significativa en sus varianzas)
### Si se acepta H[0] y se rechaza H[1] se hace la prueba t, de lo contrario se hace una prueba U Mann Whitney.
## Prueba T:
### H[0] - La media de las muestras no tiene diferencias significativas.
### H[1] - La media de las muestras tiene diferencias significativas.
## Prueba U Mann Withney:
### H[0] - Las medianas de las muestras no tienen diferencias significativas.
### H[1] - Las medianas de las muestras tienen diferencias significativas.

```python
print("Pruebas Estadísticas Población Control\n")
p_value_levene=sis.levene(ana_c['Canal 1'],ana_c['Canal 2'],ana_c['Canal 3'],ana_c['Canal 4']
                        ,ana_c['Canal 5'],ana_c['Canal 6'],ana_c['Canal 7'],ana_c['Canal 8'])[1]
print('Valor p de la prueba de Levene:', p_value_levene)
if p_value_levene >= 0.05:
    t_statistic, p_value_t=sis.ttest_ind(ana_c['Canal 1'],ana_c['Canal 2'],ana_c['Canal 3'],ana_c['Canal 4']
                        ,ana_c['Canal 5'],ana_c['Canal 6'],ana_c['Canal 7'],ana_c['Canal 8'])
    # Interpretar los resultados de la prueba t
    if p_value_t < 0.05:
        print('Se encontraron diferencias estadísticamente significativas entre los grupos.')
    # Identificar los canales que proporcionan información diferencial
    else:
        print('No se encontraron diferencias estadísticamente significativas entre los grupos.')
else:
    # Realizar la prueba U de Mann-Whitney si no se cumplen los supuestos para la prueba t
    concatenada1=ana_c['Canal 1']
    concatenada2=np.concatenate((ana_c['Canal 2'],ana_c['Canal 3'],ana_c['Canal 4']
                        ,ana_c['Canal 5'],ana_c['Canal 6'],ana_c['Canal 7'],ana_c['Canal 8']))
    u_statistic, p_value_mw = sis.mannwhitneyu(concatenada1,concatenada2)
    # Interpretar los resultados de la prueba U de Mann-Whitney
    if p_value_mw < 0.05:
        print('Se encontraron diferencias estadísticamente significativas entre los grupos.')
    # Identificar los canales que proporcionan información diferencial
    else:
        print('No se encontraron diferencias estadísticamente significativas entre los grupos.')
```
