```python
#Se importa la función Numpy, la cual es objeto de estudio de este taller.
import numpy as np
```

```python
#Punto a. Creación de vectores.
a = np.array([67.1,1,-0.3,5.2,-6])
b = np.array([1,3,2.2,5.1,1])
print('a =',a)
print('b =',b)
```

```python
#Punto b. Multiplicación escalar de vectores (producto punto).
result = np.dot(a,b)
print('(a∙b) =',result)
```

```python
#Punto c. Multiplicación punto a punto de vectores.
cross = a*b
print(cross)
```

```python
#Punto d. Construcción de una matriz.
A = np.array([[2,-1,-3], [4,-1.5,-2.5], [7.3,-0.9,-0.2]])
print('Matriz A =',A)
```

```python
#Punto e. Traspuesta de una matriz.
A_Tras = A.T
print('Matriz traspuesta de A =',A_Tras)
```

```python
#Punto f. Funciones de la librería numpy.
#Función 'ones': devuelve una matriz de n*m elementos, los cuales van a ser 1.
n = int(input('Ingrese el número de filas de la matriz: '))
m = int(input('Ingrese el número de columnas de la matriz: '))
ones = np.ones((n, m))
print(ones)
#En este caso, se puede asignar los valores de las filas y las columnas, por ejemplo, 3 filas y 2 columnas.
```

```python
#Función 'round': redondea los elementos de una matriz a un número determinado de decimales.
B = np.array([[1.344952,6.7435],[3.141592653,-0.15395]])
deci = int(input('Ingrese el número de decimales que desea obtener: '))
B_round = np.round(B,decimals=deci)
print('Matriz B =',B)
print('Matriz B redondeada =',B_round)
#Los elementos de la matriz B se han redondeado al número de decimales ingresado.
```

```python
#Función 'ceil': similar a la función round, redondea los elementos de una matriz o un arreglo al número entero inmediatamente mayor a este.
C = np.array([[1.344952,6.7435],[3.141592653,-1.15395]])
C_round = np.ceil(C)
print('Matriz C =',C)
print('Matriz C redondeada =',C_round)
#Los elementos de la matriz C se han redondeado al número de decimales ingresado.
```

```python
#Función 'floor': contrario a la función ceil, la función floor redondea los elementos de una matriz al entero inmediatamente menor a este.
D = np.array([[1.344952,6.7435],[3.141592653,-1.15395]])
D_round = np.floor(D)
print('Matriz D =',D)
print('Matriz D redondeada =',D_round)
#Los elementos de la matriz C se han redondeado al número de decimales ingresado.
```

```python
#Punto g. Acceso a un elemento específico de una matriz. Se desea acceder al elemento de la n-ésima fila (1 para este caso) y m-ésima columna (3 para este caso), pero como la indexación
#comienza desde cero (0), los argumentos que se  dan a este elemento son n-1 (0) y m-1 (2).
element = A[0,2]
print(int(element))
```

```python
#Punto h. Acceso a una fila o columna de una matriz. La indexación se hace de igual manera que el literal g. Los dos puntos (:) indican que se toman todos los elementos de ese argumento.
row = A[1,:]
print(row)
```

```python
#Punto i. Dimensiones de una matriz: se tiene los comandos 'dim' y 'shape', los cuales entregan el número de dimensiones y las dimensiones (tupla) de una matriz.
num_dim = A.ndim
print(num_dim)
dim = A.shape
print(dim)
```

```python
#Punto j. Función y1(n) = sin(π*0.18*n) para 0 ≤ n ≤ 80.
#Primero es necesario establecer el  dominio de la función, recordando que se requieren 3 argumentos: inicio, fin y paso (al cual llamaremos k, para este caso), el inicio se establece en cero (0)
#por defecto, pero el fin tiene una particularidad debido a la indexación, si por ejemplo, el argumento de fin es 80, el último elemento del dominio va a ser 80 - k, ya que la indexación
#comienza desde 0.
dom = np.arange(0,80.5,0.5)
#print(dom)
#Se establece un paso de 0.5 con el fin de obtener más puntos de evaluación para la función que se va a construir, pero la elección queda a criterio del programador.
y1 = np.sin(np.pi*0.18*dom)
print(y1)
```

```python
#Punto k. Función y2(n) = cos(2π*0.03*n) en el mismo intervalo.
y2 = np.cos(2*np.pi*0.03*dom)
#print(y2)
```

```python
#Punto l. Suma y producto de las funciones anteriores.
suma=y1+y2
multi = y1*y2
```

```python
#Punto m. Gráficas de las señales y1 e y2.
#Para graficar es necesario importar la librería Matplot.
import matplotlib.pyplot as plt
plt.subplot(2,1,1)
plt.plot(dom,y1,label = 'Señal y1',linestyle=':',linewidth=1)
plt.plot(dom,y2,label = 'Señal y2',linestyle='-.',linewidth=2.5)
plt.xlabel('Dominio de la función (n)')
plt.ylabel('Amplitud')
plt.title('Señales iniciales')
plt.legend()
plt.show()
```

```python
#Punto n. Graficar las funciones resultantes de sumar y multiplicar y1 con y2.
plt.subplot(2,1,2)
plt.plot(dom,suma,color='red',label = 'Señal suma')
plt.plot(dom,multi,color='green',label = 'Señal producto',linestyle='dashed',linewidth=1.3)
plt.xlabel('Dominio de la función (n)')
plt.ylabel('Amplitud')
plt.title('Señales iniciales')
plt.legend()
plt.show()
```

```python
#RETO.
#Para convertir el diccionario a un dataframe, es necesario importar otra librería, en este caso, Pandas.
import pandas as pd
notas ={}
def notas_lab(notas):
    #Convertir el diccionario de notas a una serie.
    notas_serie = pd.Series(notas)

    # Calcular la nota mínima, máxima y media.
    min = notas_serie.min()
    max = notas_serie.max()
    media = notas_serie.mean()

    # Calcular la desviación típica utilizando numpy.
    desv_est = np.std(notas_serie)

    # Crear una serie con los resultados.
    stat = pd.Series({'Nota mínima': min, 'Nota máxima': max, 'Nota media': media, 'Desviación típica': desv_est})

    return stat

#Ahora, este ciclo permite introducir el número de documento del estudiante y la calificación asignada, si desea añadir otro estudiante, lo puede hacer, de lo contrario, se muestra los valores
#solicitadaos.
while True:
  doc = input('Ingrese el documento del estudiante: ')
  calificacion = float(input('Ingrese la calificación: '))
  notas[doc]=calificacion
  op = input('¿Desea agregar otro estudiante? Responda (SI) o (NO)')
  if op == 'NO':
    break
resultados = notas_lab(notas)
print(resultados)
```
