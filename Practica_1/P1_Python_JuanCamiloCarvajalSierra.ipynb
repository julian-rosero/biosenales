```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Punto a
a=np.array([67.1,1,-0.3,5.2,6])
b=np.array([1,3,2.2,5.1,1])
```

```python
#Punto d
A=np.array([[2,-1,-3],[4,-1.5,-2.5],[7.3,-0.9,-0.2]])
print(A)
```

```python
#Punto f.ones
# ones es una función que crea una matriz de unos de tamaño nxm. Recibe datos en forma (n,m)
filas=int(input('Filas: '))
columnas=int(input('Columnas: '))
max=np.ones((filas,columnas))
print(max)
```

```python
#Punto f.round
# round es una función que aproxima los elementos de un array a números decimales con decimales determinados
max=np.array([[1.234, 4.567], [7.890, 4.125]])
max_rnd=np.round(max,1)
print(max_rnd)
```

```python
# Punto f.ceil
# ceil es una función que devuelve el entero mayor más cercano al elemento en un array
max_cl=np.ceil(max)
print(max_cl)
```

```python
# Punto f.floor
# floor es una función que devuelve el entero menor más cercano al elemento en un array
min_fl=np.floor(max)
print(min_fl)
```

```python
#Punto j y k
intervalo=np.arange(1,81,0.001)
y1=np.sin(np.pi*0.18*intervalo)
y2=np.cos(2*np.pi*0.03*intervalo)
while True:
  n=int(input('Ingrese un valor entre 0 y 80: '))
  if n<0 or n>80:
    print('Fuera de Rango')
  else:
    print('\n')
    print('Y1('+str(n)+')=Sin(pi*0.18*'+str(n)+') = '+str(y1[n-1]))
    print('\n')
    print('Y2('+str(n)+')=Cos(2*pi*0.03*'+str(n)+') = '+str(y2[n-1]))
    break
```

```python
#Punto l
s=y1+y2
t=y1*y2
print('Señal 3 (Y1+Y2)\n')
print(s)
print('\nSeñal 4 (Y1.Y2)\n')
print(t)
```

```python
#Punto m
plt.plot(intervalo,y1,label='Y1',color='black')
plt.plot(intervalo,y2,label='Y2',color='red')
plt.xlabel('Intervalo')
plt.ylabel('Valores')
plt.title('Y1 vs Y2')
plt.legend()
plt.show()
```

```python
##Punto n
plt.plot(intervalo,s,label='S',color='blue')
plt.plot(intervalo,t,label='T',color='green')
plt.xlabel('Intervalo')
plt.ylabel('Valores')
plt.title('S vs T')
plt.legend()
plt.show()
```
